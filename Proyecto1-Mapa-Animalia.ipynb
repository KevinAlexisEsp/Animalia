```python

with open('Lista_urls.txt', 'r') as archivo:
    # Lee todas las líneas del archivo y elimina los espacios en blanco y caracteres de nueva línea
    lista_url = [line.strip() for line in archivo.readlines()]
```

```python
nombre_animal = []
paises = []
paises_animal = []
paises_limpios = []
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}


for url in tqdm(lista_url[:]):
    response = requests.get(url, headers=headers)
    print(url)
    
    soup = BeautifulSoup(response.text, "html.parser")
    try:
        animal = soup.find("h1", class_ = "a-h1").text.replace("\n","")
        nombre_animal.append(animal)
    except:
        nombre_animal.append(np.nan)
    #     DATOOS ANIAMLES(KINGDOM, PHYLUM, SUBPHYYLUM, CLASS, ORDER, SUBORDER, FAMILY, GENUS)
    
    verificador_paises = soup.find_all("div", class_="s-distr-geography__slug")
    for num, pais in enumerate(verificador_paises):
        pais = pais.text.replace("\n","")
        try:
            if pais == "Countries":
                verificador = True
                for a in soup.find_all("div", class_="col-sm-9")[num]:  
                    if a.text.replace("\n","") != "Show More" and a.text.replace("\n","") != "Show Less" and verificador == True:
                        paises.append(a.text.replace("\n","")) 
                    else:
                        verificador = False
                for a in soup.find_all("span", class_="read-more-content"):
                    if a.text != "Show Less":
                        elemento = (a.text.replace("\n","").split(","))
                        for n in elemento:
                            paises.append(n)

                paises_limpios = [pais.strip(', ') for pais in paises if pais.strip(', ')] 
                try:
                    paises_limpios[-1] = paises_limpios[-1].split(" ")[0]
                except:
                    pass
        except:
            pass
        paises_animal.append(paises_limpios) 
        paises = []
```
